### Load Sample Data

```python
import pandas as pd
sample_data = pd.read_csv('../../data/games.csv', nrows=5)
sample_data.drop('moves', axis=1)
```

### Removing a required column to test error handling

```python
test_missing_columns = sample_data.drop(['opening_name'], axis=1)
test_missing_columns.drop('moves', axis=1).head()
```

```python
test_missing_columns.to_csv('missingColumns.csv', index=False)
```

### Creating a subset with invalid data

```python
test_invalid_data = sample_data.copy()
test_invalid_data.loc[0, 'opening_name'] = 12345  # Assigning a number to a string column
test_invalid_data.drop('moves', axis=1).head()
```

```python
test_invalid_data.to_csv('invalidData.csv', index=False)
```

### Creating a subset with valid data

```python
test_valid_data = sample_data
test_valid_data.drop('moves', axis=1)
```

```python
test_valid_data.to_csv('validData.csv', index=False)
```
